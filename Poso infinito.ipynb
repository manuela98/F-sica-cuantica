## Poso de Potencial infinito simétrico  

En este caso el potencial esta dado por  
$V(X) = 0 $  si |x| >L/2  
$V(X) = ∞ $  |x|>= L/2

```python
import numpy as np 
import matplotlib.pylab as plt

# Funcion de onda con energia cuantizada 

# Soluciones para con coseno 
l = 100.
x = np.linspace(-l/2.,l/2,100)
def fi(x):
    
   
    N = 1.
    hbar2 = 1.
    m = 1.
    n = 1. # Con n impar
    # Minima energia con n = 1 
    E = (hbar2*np.pi**2*n**2)/(2*m*l**2)
    
    return N*np.cos((np.sqrt(2*m*E)/hbar2)*x)

f= fi(x)
plt.plot(x,f)
plt.show()
```

```python
# Soluciones para el seno

def fi2(x):
    
   
    N = 1.
    hbar2 = 1.
    m = 1.
    n = 4. # Con n pares 
    # Minima energia con n = 1 
    E = (hbar2*np.pi**2*n**2)/(2*m*l**2)
    
    return N*np.sin((np.sqrt(2*m*E)/hbar2)*x)

f= fi2(x)
plt.plot(x,f)
plt.show()
```

```python
# Probabilidad de encontrar la funcion de onda en un lugar abs(fi)**2

l = 100.
x = np.linspace(-l/2.,l/2,100)
def abfi(x):
    
   
    N = 1.
    h2 = 1.
    m = 1.
    n = 2.
    E = ((2*np.pi**2*h2)/(m*l**2))*h2*n**2
    
    return abs(N*np.sin((np.sqrt(2*m*E)/h2)*x))**2

f= abfi(x)
plt.plot(x,f)
```

```python
# Rompiendo la simetria del problema 

# de 0 a l 

l = 100.
x = np.linspace(0,l,100)
def fi(x):
    
   
    N = 1.
    h2 = 1.
    m = 1.
    n = 1.
    # Minima energia con n = 1 
    E = ((2*np.pi**2*h2)/(m*l**2))*h2*n**2
    
    return N*np.sin((np.sqrt(2*m*E)/h2)*x)

f= fi(x)
plt.plot(x,f)
```
