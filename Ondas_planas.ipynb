#  Metodo  Variacional 

La  ecuación  de   Schrödinger  tiene  la  forma 

$-\frac{\hbar^2}{2m}\nabla^2\psi(x)+V(x)\psi(x)=E\psi(x)$
en una notación  más  compacta 

$\hat{H}|\psi>=E|\psi>$,  para   calcular la  energia  calcularemos  el  valor  esperado  del  hamiltoniano 
$<\psi| \hat{H}|\psi>=<\psi|E|\psi>$

$<\psi| \hat{H}|\psi>=E<\psi|\psi>$, finalmente  

$E=\frac{<\psi| \hat{H}|\psi>}{<\psi|\psi>}$

la  función de   onda  la  escribiremos como:

$\psi(x)=\sum_{n}C_{n}e^{iG_{n}x}$


$\frac{\hbar^2}{2m}\sum_{n}C_{n}G_{n}^{2}e^{i(G_n-G_m)x}+\sum_{n}C_{n}V(x)e^{i(G_{n}-G_{m})x}=E\sum_{n}C_{n}e^{i(G_{n}-G_{m})x}$

integrando  en  todo  el  espacio  tenemos 

$\frac{\hbar^2}{2m}\sum_{n}C_{n}G_{n}^2L\delta_{nm}+\sum_{n}C_{n}\int V(x)e^{i(G_{n}-G_{m})x}dx=E\sum_{n} C_{n}L\delta_{nm}$ 


dividiendo  por  L

$\frac{\hbar^2}{2m}\sum_{n}C_{n}G_{n}^2\delta_{nm}+\frac{1}{L}\sum_{n}C_{n}\int_{-L/2}^{L/2} V(x)e^{i(G_{n}-G_{m})x}dx=E\sum_{n} C_{n}\delta_{nm}$
