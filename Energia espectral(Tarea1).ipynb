```python
# Densidad de energía espectral pag 21 

import numpy as np
from math import pi
import matplotlib.pylab as plt

# Temperatura [K] 

# Frecuencia [Revoluciones/S]

v = np.linspace(1e8,1e15,3000)

def U(v,T):
    
    #constants 
    # Cte Boltzmann [J/K]
    k = 1.38*10**-23
    # Cte de Plank [J*S]
    h = 6.6260*10**-34
    # Velocidad de la luz [k/s]
    c = 300000
    return (8*pi/c**3)*h*v**3/(np.exp(h*v/(k*T))-1)
u = []

for i in range(3000):
    u += [U(v[i],T)]
T = 4000
plt.plot(v,u)
    
```
